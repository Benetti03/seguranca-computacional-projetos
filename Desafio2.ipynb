```python
#Como há um único token que não foi possível decodificar no site, considera-se esse o token vulnerável.

def decode_base64url(encoded_str):
    padding = '=' * (4-len(encoded_str) % 4)
    return base64.urlsafe_b64decode(encoded_str + padding).decode('utf-8')

header, payload, signature = undecoded_token.split(".")

decoded_header = decode_base64url(header)
decoded_payload = decode_base64url(payload)
print("Decoded Header:", decoded_header)
print("Decoded Payload:", decoded_payload)
print("FLAG DO TOKEN VULNERÁVEL:", json.loads(decoded_payload)["flag"])
```

```python
#Baseado no header decodificado, percebe-se que esse "undecoded_token" é o vulnerável, pois o algoritmo de assinatura é "none", ou seja, não há assinatura.
#O payload decodificado contém a flag "elefante". Portanto, a flag do token vulnerável é "elefante".
```
